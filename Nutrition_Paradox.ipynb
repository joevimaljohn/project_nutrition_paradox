```python
import pycountry

# Special mapping for WHO / WB region codes
special_cases = {
    'GLOBAL': 'Global',
    'WPR': 'Western Pacific Region',
    'SEAR': 'South-East Asia Region',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'Europe',
    'AFR': 'Africa',
    'AMR': 'Americas Region',
    'WB_HI': 'High Income',
    'WB_LMI': 'Low & Middle Income',
    'WB_LI': 'Low Income',
    'WB_UMI': 'Upper Middle Income'
}

# Function to convert codes to names
def code_to_country_name(code):
    code = str(code).strip().upper()
    if code in special_cases:
        return special_cases[code]
    country = pycountry.countries.get(alpha_3=code)
    return country.name if country else code

# Apply safely without SettingWithCopyWarning
df_obesity = df_obesity.copy()
df_malnutrition = df_malnutrition.copy()

df_obesity['Country'] = df_obesity['Country'].apply(code_to_country_name)
df_malnutrition['Country'] = df_malnutrition['Country'].apply(code_to_country_name)
```

Top 5 countries least reliable countries(with highest CI_Width) and Top 5 most consistent countries (smallest average CI_Width)


```python
query = """
SELECT 'Least Reliable' AS Category, Country, Avg_CI_Width
FROM (
    SELECT Country, ROUND(AVG(CI_Width), 2) AS Avg_CI_Width
    FROM obesity
    GROUP BY Country
    ORDER BY Avg_CI_Width DESC
    LIMIT 5
) AS least

UNION ALL

SELECT 'Most Consistent' AS Category, Country, Avg_CI_Width
FROM (
    SELECT Country, ROUND(AVG(CI_Width), 2) AS Avg_CI_Width
    FROM obesity
    GROUP BY Country
    ORDER BY Avg_CI_Width ASC
    LIMIT 5
) AS most;
"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

Top 10 countries with consistent low obesity (low avg + low CI)

```python
query = """
SELECT Country,
       ROUND(AVG(Mean_Estimate),2) AS Avg_Obesity,
       ROUND(AVG(CI_Width),2)     AS Avg_CI
FROM obesity
GROUP BY Country
HAVING Avg_Obesity < 15 AND Avg_CI < 2
ORDER BY Avg_Obesity ASC, Avg_CI ASC
LIMIT 10;

"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

Countries where female obesity exceeds male by large margin (same year)

```python
query = """
SELECT f.Country, f.Year,
       ROUND(f.Mean_Estimate,2) AS Female_Obesity,
       ROUND(m.Mean_Estimate,2) AS Male_Obesity,
       ROUND(f.Mean_Estimate - m.Mean_Estimate,2) AS Difference
FROM obesity f
JOIN obesity m
  ON f.Country = m.Country AND f.Year = m.Year
WHERE f.Gender = 'Female' AND m.Gender = 'Male'
  AND (f.Mean_Estimate - m.Mean_Estimate) > 5
ORDER BY Difference DESC;

"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

Countries with increasing malnutrition (trend detection)

```python
query = """
SELECT
    Country,
    MIN(Mean_Estimate) AS Min_Malnutrition,
    MAX(Mean_Estimate) AS Max_Malnutrition,
    (MAX(Mean_Estimate) - MIN(Mean_Estimate)) AS Changes
FROM malnutrition
GROUP BY Country
HAVING (MAX(Mean_Estimate) - MIN(Mean_Estimate)) > 0
ORDER BY (MAX(Mean_Estimate) - MIN(Mean_Estimate)) DESC;

"""
cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

Gender-based disparity in both obesity and malnutrition


```python
query = """
SELECT
    o.Gender,
    ROUND(AVG(o.Mean_Estimate), 2) AS Avg_Obesity,
    ROUND(AVG(m.Mean_Estimate), 2) AS Avg_Malnutrition,
    ROUND(AVG(o.Mean_Estimate) - AVG(m.Mean_Estimate), 2) AS Obesity_Malnutrition_Diff
FROM obesity o
JOIN malnutrition m
    ON o.Country = m.Country
    AND o.Year = m.Year
    AND o.Gender = m.Gender
GROUP BY o.Gender
ORDER BY o.Gender;
"""
cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

Region-wise avg estimates side-by-side(Africa and America)


```python
query = """
SELECT
    Year,
    ROUND(AVG(CASE WHEN Region = 'Africa' THEN Mean_Estimate END), 2) AS Africa_Avg,
    ROUND(AVG(CASE WHEN Region = 'Americas' THEN Mean_Estimate END), 2) AS America_Avg
FROM obesity
WHERE Region IN ('Africa', 'Americas')
GROUP BY Year
ORDER BY Year;
"""
cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

Countries with obesity up & malnutrition down


```python
query = """
SELECT
    o.Country,
    (MAX(o.Mean_Estimate) - MIN(o.Mean_Estimate)) AS Obesity_Change,
    (MIN(m.Mean_Estimate) - MAX(m.Mean_Estimate)) AS Malnutrition_Change
FROM obesity o
JOIN malnutrition m
    ON o.Country = m.Country
    AND o.Year = m.Year
GROUP BY o.Country
HAVING
    (MAX(o.Mean_Estimate) - MIN(o.Mean_Estimate)) > 1
    AND (MAX(m.Mean_Estimate) - MIN(m.Mean_Estimate)) < 1
ORDER BY Obesity_Change DESC;
"""
cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```
